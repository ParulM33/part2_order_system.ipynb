```

print("MENU BY CATEGORY")

categories = []
# Collect categories manually (could have used set but keeping simple)
for item in menu:
    cat = menu[item]["category"]
    if cat not in categories:
        categories.append(cat)

for cat in categories:
    print("
=====", cat, "=====")
    for item in menu:
        if menu[item]["category"] == cat:
            status = "Available"
            if menu[item]["available"] == False:
                status = "Unavailable"
            print(item, "₹", menu[item]["price"], "[", status, "]")

# Simple statistics
count_all = len(menu)
available_items = 0
for item in menu:
    if menu[item]["available"]:
        available_items += 1

highest = 0
expensive_item = ""
for item in menu:
    if menu[item]["price"] > highest:
        highest = menu[item]["price"]
        expensive_item = item

print("
Total items:", count_all)
print("Available items:", available_items)
print("Most expensive:", expensive_item, highest)
print("Items under 150:")
for item in menu:
    if menu[item]["price"] < 150:
        print(item, menu[item]["price"])
```

```

cart = []

def add_item(name, qty):
    # Check presence in menu
    if name not in menu:
        print(name, "not in menu")
        return
    if not menu[name]["available"]:
        print(name, "is unavailable")
        return
    for entry in cart:
        if entry["item"] == name:
            entry["quantity"] += qty
            # Forgot return earlier and got duplicates
            print("Quantity updated for", name)
            return
    cart.append({"item": name, "quantity": qty, "price": menu[name]["price"]})
    print("Added", name)

def remove_item(name):
    for entry in cart:
        if entry["item"] == name:
            cart.remove(entry)
            print("Removed", name)
            return
    print(name, "not found in cart")
```

```

add_item("Paneer Tikka", 2)
add_item("Gulab Jamun", 1)
add_item("Paneer Tikka", 1)
add_item("Mystery Burger", 1)
add_item("Chicken Wings", 1)
remove_item("Gulab Jamun")

print("
ORDER SUMMARY")
subtotal = 0
for entry in cart:
    total = entry["quantity"] * entry["price"]
    subtotal += total
    print(entry["item"], "x", entry["quantity"], "=", total)

gst = subtotal * 0.05
print("Subtotal:", subtotal)
print("GST:", gst)
print("Total:", subtotal + gst)
```

```

import copy

# First attempt was inventory_backup = inventory (wrong)
inventory_backup = copy.deepcopy(inventory)

inventory["Paneer Tikka"]["stock"] = 0
print("Inventory:", inventory["Paneer Tikka"])
print("Backup:", inventory_backup["Paneer Tikka"])

# Restore inventory
inventory = copy.deepcopy(inventory_backup)

for entry in cart:
    item = entry["item"]
    qty = entry["quantity"]
    if inventory[item]["stock"] >= qty:
        inventory[item]["stock"] -= qty
    else:
        inventory[item]["stock"] = 0

print("
Reorder Alerts")
for item in inventory:
    if inventory[item]["stock"] <= inventory[item]["reorder_level"]:
        print("Reorder", item, "only", inventory[item]["stock"], "left")
```

```

print("
DAILY SALES")
daily = {}
for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]
    daily[date] = total
    print(date, total)

best_day = ""
best_value = 0
for day in daily:
    if daily[day] > best_value:
        best_day = day
        best_value = daily[day]
print("Best day:", best_day)

item_count = {}
for date in sales_log:
    for order in sales_log[date]:
        for it in order["items"]:
            if it not in item_count:
                item_count[it] = 1
            else:
                item_count[it] += 1

print("Most ordered item:", max(item_count, key=item_count.get))

# Add new day
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0}
]

print("
ALL ORDERS")
count = 1
for date in sales_log:
    for order in sales_log[date]:
        print(count, date, order["order_id"], order["items"], order["total"])
        count += 1
```
